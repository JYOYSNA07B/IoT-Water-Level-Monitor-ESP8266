# **IoT-Water-Level-Monitor-ESP8266**

```python
#define BLYNK_PRINT Serial
#define BLYNK_TEMPLATE_ID "TMPL3AJLCDD0I"
#define BLYNK_TEMPLATE_NAME "water tank control using IOT"
#define BLYNK_AUTH_TOKEN "PkVe5HzOv-KfmgWaX49r2S4vbmSAlIud"
char auth[] = BLYNK_AUTH_TOKEN;
char ssid[] = "adminn";
char pass[] = "Password0";
#include <ESP8266WiFi.h>
#include <BlynkSimpleEsp8266.h>
WidgetLED led1(V2);
const int motor = D4;
const int trigPin = D5;
const int echoPin = D6;
int distance = 0;
unsigned int keep = 30;
BlynkTimer timer;
void setup()
{
  Serial.begin(9600);
  Blynk.begin(auth, ssid, pass);
  pinMode(trigPin, OUTPUT); // Sets the trigPin as an Output
  pinMode(echoPin, INPUT); // Sets the echoPin as an Input
  pinMode(motor, OUTPUT);
digitalWrite(motor, LOW);
delay(300);
}
void US()
{
  ////////////////////////////////////////////////////////////////Ultrasonic sensor
  digitalWrite(trigPin, LOW);
  delayMicroseconds(2);
  digitalWrite(trigPin, HIGH);
  delayMicroseconds(2);
  digitalWrite(trigPin, LOW);
  long timedelay = pulseIn(echoPin, HIGH);
  distance = 0.0343 * (timedelay / 2);
  distance = (keep - distance);
  Serial.print("Distance :");
  Serial.println(distance);
  if (distance < 5)// LOW edit
  {
    led1.on();
    digitalWrite(motor, HIGH);
  }
  if (distance > 25)// HI deit
  {
  led1.off();
  digitalWrite(motor, LOW);
  }
}
void loop()
{
  Blynk.run();
  timer.run();
  US();delay(500);
  Blynk.virtualWrite(V1, distance); delay(20);
}
```
